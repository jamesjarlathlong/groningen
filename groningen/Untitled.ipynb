```python
import tensorflow as tf
import functools
```

```python
def label(i):
    return 0 if i<10 else 1
def gen(nn):
    ii = 0
    while ii < nn:
        ii += 1
        yield ii, ii+4, label(ii*2)
def input_fn(n):
    g = lambda: gen(n)
    ds = tf.data.Dataset.from_generator(g, (tf.int64, tf.int64, tf.int64)) 
    ds = ds.batch(1)
    x, y, z = ds.make_one_shot_iterator().get_next()
    print(x,y)
    return {'x': x, 'y': y}, z

def build_columns():
    x = tf.feature_column.numeric_column('x')
    y = tf.feature_column.numeric_column('y')
    return [x, y]

def mainr():
    model = tf.estimator.DNNClassifier(
        feature_columns=build_columns(),
        # The model must choose between 4 classes
        n_classes=2,
        # Two hidden layers of 100 nodes each.
        hidden_units=[100, 100]
        )
    model.train(input_fn=lambda:input_fn(100))
    results = model.evaluate(input_fn = lambda:input_fn(20))
    print(results)
```

```python
mainr()
```
